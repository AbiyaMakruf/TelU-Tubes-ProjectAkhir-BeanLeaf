# Import Library

```python
import os
import tensorflow as tf
import pandas as pd
import numpy as np
import time
import csv
import os
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

# Mengubah model .keras menjadi .tflite

```python
def convert_keras_to_tflite(input_dir, output_dir):
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Get all .keras files in the input directory
    keras_files = [f for f in os.listdir(input_dir) if f.endswith('.keras')]

    if not keras_files:
        print("No .keras files found in the input directory.")
        return

    for keras_file in keras_files:
        try:
            # Load the .keras model
            keras_model_path = os.path.join(input_dir, keras_file)
            model = tf.keras.models.load_model(keras_model_path)

            # Convert the model to TFLite
            converter = tf.lite.TFLiteConverter.from_keras_model(model)
            tflite_model = converter.convert()

            # Save the TFLite model
            tflite_file_name = os.path.splitext(keras_file)[0] + ".tflite"
            tflite_model_path = os.path.join(output_dir, tflite_file_name)
            with open(tflite_model_path, "wb") as f:
                f.write(tflite_model)

            print(f"Successfully converted {keras_file} to {tflite_file_name}")

        except Exception as e:
            print(f"Failed to convert {keras_file}: {e}")

if __name__ == "__main__":
    # Prompt user for input and output directories
    input_dir = "results/best_model/"
    output_dir = "results/best_model/tflite"

    # Perform conversion
    convert_keras_to_tflite(input_dir, output_dir)
```

# Membuat tabel comparison file size

```python
def generate_file_size_table(keras_dir, tflite_dir, output_csv):
    # Get all .keras and .tflite files
    keras_files = [f for f in os.listdir(keras_dir) if f.endswith('.keras')]
    tflite_files = [f for f in os.listdir(tflite_dir) if f.endswith('.tflite')]

    data = []

    for index, keras_file in enumerate(keras_files, start=1):
        keras_file_path = os.path.join(keras_dir, keras_file)
        keras_size = os.path.getsize(keras_file_path) / (1024 * 1024)  # Size in MB

        tflite_file = os.path.splitext(keras_file)[0] + ".tflite"
        tflite_file_path = os.path.join(tflite_dir, tflite_file)
        tflite_size = os.path.getsize(tflite_file_path) / (1024 * 1024) if os.path.exists(tflite_file_path) else None

        data.append([keras_file, round(keras_size, 2), round(tflite_size, 2) if tflite_size else None])

    # Create a DataFrame
    df = pd.DataFrame(data, columns=["File Name (.keras)", "Size (.keras) MB", "Size (.tflite) MB"])

    # Save to CSV without index
    df.to_csv(output_csv, index=False)
    print(f"Table saved to {output_csv}")

keras_dir = "results/best_model/"
tflite_dir = "results/best_model/tflite"
output_csv = "results/csv/tabel_comparison_keras_tflite.csv"

generate_file_size_table(keras_dir, tflite_dir, output_csv)
```

# Membuat tabel inference time

```python
def load_tflite_model(tflite_model_path):
    # Load TFLite model and allocate tensors
    interpreter = tf.lite.Interpreter(model_path=tflite_model_path)
    interpreter.allocate_tensors()
    return interpreter

def run_keras_inference(model_path, input_data):
    # Load Keras model
    model = tf.keras.models.load_model(model_path)
    start_time = time.time()
    predictions = model.predict(input_data)
    end_time = time.time()
    inference_time = end_time - start_time
    return predictions, inference_time

def run_tflite_inference(interpreter, input_data):
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    # Ensure input data matches the input tensor's expected shape and type
    input_data = np.array(input_data, dtype=input_details[0]['dtype'])

    interpreter.set_tensor(input_details[0]['index'], input_data)
    start_time = time.time()
    interpreter.invoke()
    end_time = time.time()

    predictions = interpreter.get_tensor(output_details[0]['index'])
    inference_time = end_time - start_time
    return predictions, inference_time

def load_test_image(image_path, target_size, preprocessing_function):
    # Load and preprocess a single image
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = preprocessing_function(img_array)  # Apply preprocessing function
    return img_array

def compare_inference(keras_dir, tflite_dir, output_csv_path, input_shapes, test_images, preprocessing_function):
    # Define class mapping
    class_mapping = {0: 'angular_leaf_spot', 1: 'bean_rust', 2: 'healthy'}

    # Get all model filenames
    keras_files = sorted([f for f in os.listdir(keras_dir) if f.endswith('.keras')])
    tflite_files = sorted([f for f in os.listdir(tflite_dir) if f.endswith('.tflite')])

    if len(keras_files) != len(tflite_files):
        raise ValueError("Number of .keras and .tflite models must be the same.")

    if len(test_images) < len(keras_files):
        additional_images = random.choices(test_images, k=len(keras_files) - len(test_images))
        test_images.extend(additional_images)

    results = []

    for keras_file, tflite_file, test_image_path in zip(keras_files, tflite_files, test_images):
        keras_model_path = os.path.join(keras_dir, keras_file)
        tflite_model_path = os.path.join(tflite_dir, tflite_file)

        # Determine input shape based on model name
        for key, target_size in input_shapes.items():
            if key in keras_file:
                input_size = target_size
                break
        else:
            raise ValueError(f"No matching input shape found for model: {keras_file}")

        # Load and preprocess the test image
        input_data = load_test_image(test_image_path, target_size=input_size, preprocessing_function=preprocessing_function)

        # Run inference using Keras model
        keras_predictions, keras_time = run_keras_inference(keras_model_path, input_data)

        # Run inference using TFLite model
        interpreter = load_tflite_model(tflite_model_path)
        tflite_predictions, tflite_time = run_tflite_inference(interpreter, input_data)

        # Get class predictions
        keras_class_index = np.argmax(keras_predictions[0])
        tflite_class_index = np.argmax(tflite_predictions[0])
        keras_class = class_mapping[keras_class_index]
        tflite_class = class_mapping[tflite_class_index]

        # Save results for this model and image
        results.append([
            os.path.basename(test_image_path), keras_file, tflite_file, keras_time, tflite_time,
            keras_class, tflite_class, keras_predictions[0].tolist(), tflite_predictions[0].tolist()
        ])

    # Save results to CSV
    with open(output_csv_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Filename", "Keras Model", "TFLite Model", "Keras Inference Time (s)", "TFLite Inference Time (s)", "Keras Class", "TFLite Class", "Keras Predictions", "TFLite Predictions"])
        writer.writerows(results)

    print(f"Results saved to {output_csv_path}")

# Example usage
if __name__ == "__main__":
    keras_dir = "results/best_model/"
    tflite_dir = "results/best_model/tflite/"
    output_csv_path = "results/csv/inference_time.csv"

    # Define input shapes for different models
    input_shapes = {
        "MobileNetV3Small": (224, 224),
        "MobileNetV3Large": (224, 224),
        "EfficientNetV2B0": (224, 224),
        "EfficientNetV2B1": (240, 240),
        "EfficientNetV2B2": (260, 260),
        "EfficientNetV2B3": (300, 300),
        "EfficientNetV2B0Modifikasi": (224, 224),
        "EfficientNetV2B1Modifikasi": (240, 240),
        "EfficientNetV2B2Modifikasi": (260, 260),
        "EfficientNetV2B3Modifikasi": (300, 300),
    }

    # List of test images (one for each model)
    test_images = [
        "datasets/testing/healthy/healthy_val.0.jpg",
        "datasets/testing/angular_leaf_spot/angular_leaf_spot_val.0.jpg",
        "datasets/testing/bean_rust/bean_rust_val.0.jpg"
    ]

    preprocessing_function = tf.keras.applications.efficientnet_v2.preprocess_input  # Example preprocessing function

    compare_inference(keras_dir, tflite_dir, output_csv_path, input_shapes, test_images, preprocessing_function)
```
